```python
import pandas as pd
```

```python
xls = pd.ExcelFile('D:\Kadre\Kadre\APRA\gitaprarepo\Apra-Monthly-Deposits\Data\c01-1-hist.xlsx')
df = pd.read_excel(xls, sheet_name='Data', header=1)
```

```python
df_cut = df.iloc[110:]
```

```python
df_cut_columns = df_cut[['Title','Number of accounts','Number of purchases','Value of purchases']]
df_cut_columns['noacc_rate_of_change_1_month'] = df_cut_columns['Number of accounts'].diff()
df_cut_columns['nopur_rate_of_change_1_month'] = df_cut_columns['Number of purchases'].pct_change()
df_cut_columns['valpur_rate_of_change_1_month'] = df_cut_columns['Value of purchases'].pct_change()
```

```python
df_cut_columns
```

```python
df_named = df_cut_columns.rename(columns={'Title':'date'})
```

```python
import plotly_express as px
```

```python

plot = px.line(df_named,x='date',y='Number of accounts',title= 'Number of Credit/Cash Card Accounts',template='plotly_white')
plot.update_traces(line_color='#333333', line_width=2)
plot.update_layout(
    xaxis_title="Time",
    yaxis_title="Volume ('000')",
    legend_title="Legend Title",
    font=dict(
        size=12,
        color="Green"
    )
)
plot.show()
```

```python

plot = px.line(df_named,x='date',y='Number of purchases',title= 'Number of Purchases with Credit/Cash Cards',template='plotly_white')
plot.update_traces(line_color='#333333', line_width=2)
plot.update_layout(
    xaxis_title="Time",
    yaxis_title="Volume ('000')",
    legend_title="Legend Title",
    font=dict(
        size=12,
        color="Green"
    )
)
plot.show()
```

```python

plot = px.line(df_named,x='date',y='Value of purchases',title= 'Value of Purchases with Credit/Cash Cards',template='plotly_white')
plot.update_traces(line_color='#333333', line_width=2)
plot.update_layout(
    xaxis_title="Time",
    yaxis_title="Amount ($M)",
    legend_title="Legend Title",
    font=dict(
        size=12,
        color="Green"
    )
)
plot.show()
```

```python
import plotly.graph_objects as go
plot = px.line(df_named,x='date',y='noacc_rate_of_change_1_month',title= 'Change in Number of Accounts (1 Month)',template='plotly_white')
plot.update_traces(line_color='#333333', line_width=2)
plot.update_layout(
    xaxis_title="Time",
    yaxis_title="Volume Change",
    legend_title="Legend Title",
    font=dict(
        size=12,
        color="Green"
    )
)
plot.add_trace(go.Scatter(x =df_named.date, y =df_named.nopur_rate_of_change_1_month, name='Number of Purchases',mode='lines'))
plot.add_trace(go.Scatter(x =df_named.date, y =df_named.valpur_rate_of_change_1_month, name='Value of Purchases',mode='lines'))
#plot.layout.yaxis.tickformat = ',.0%'
plot.show()
```
