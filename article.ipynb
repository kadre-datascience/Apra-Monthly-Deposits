```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
def growth_plot():


    pio.templates.default = "plotly_white"
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    fig.add_trace(go.Scatter(y=debit_cut['Number of purchases'],x=debit_cut.Title,name = 'Debit Card Purchases'
    # color_discrete_map={
    #                 "Big 4": "brown",
    #                 "International retail bank": "#565656",
    #                 "Tier 2 bank (1)": "#316395",
    #                 "Tier 2 bank (2)": "rgb(15,133,84)",
    #                 "Other bank": "rgb(111,64,112)"},
    ),secondary_y=False)
    fig.update_layout(legend=dict(
        orientation="h",
        title="Card Types",
#        yanchor="bottom",
#        y=1.2,
#        xanchor="left",
 #       x=0.4)
        )
        ,title_text = "Debit vs Credit Card Purchases")
    fig.add_trace(go.Scatter(x=df_cut.Title, y=df_cut['Total number of transactions'],
                    mode='lines',
                    name='Credit Card Purchases'),secondary_y=True)
    fig.update_xaxes(title= "")
    fig.update_traces(mode='lines')
    fig.update_xaxes(title_text='Time' )
    fig.update_yaxes(title_text='Volume (Millions)' )
    fig.update_yaxes(title_text='Volume (000)',secondary_y=True )
    
    # fig.add_annotation(x='2022-08-01', y=2.81,
    #             text="+281%",
    #             showarrow=False,
    #             arrowhead=1)
    # fig.add_annotation(x='2022-08-01', y=0.35,
    #             text="+31%",
    #             showarrow=False,
    #             arrowhead=1)
    # fig.add_annotation(x='2022-08-01', y=0.97,
    #             text="+97%",
    #             showarrow=False,
    #             arrowhead=1)
    # fig.add_annotation(x='2022-08-15', y=0.11,
    #             text="Big4 +11%",
    #             showarrow=False,
    #             arrowhead=2)
    #fig.show()  
    return fig
```
