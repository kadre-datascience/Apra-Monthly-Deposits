```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>")) 

df = pd.read_excel('Data/Monthly authorised deposit-taking institution statistics back-series March 2019 - December 2022.xlsx',sheet_name= 'Table 1',header=1)
RBAdf = pd.read_excel('Data/RBAcredit.xlsx',sheet_name= 'Data',header=1)
```

```python
RBAdf_cut = RBAdf.iloc[519:]
#df.columns = RBAdf_cut.columns.str.replace(' ','_')
```

```python
RBAdf_cut['Total_Housing_Credit'] = RBAdf_cut['Credit; Owner-occupier housing; Seasonally adjusted']+RBAdf_cut['Credit; Investor housing; Seasonally adjusted']
```

```python
df.columns = df.columns.str.replace(' ','_')
df['Total_Housing_Loans'] =  df['Loans_to_households:_Housing:_Owner-occupied'] +  df['Loans_to_households:_Housing:_Investment'] 
```

```python
df
```

```python
# convert date column to datetime
df['date'] = pd.to_datetime(df['Period'])

# group by year and month
grouped = df.groupby([df['date'].dt.year, df['date'].dt.month])

# calculate sum for each group
df['Month_Sum'] = grouped['Total_Housing_Loans'].transform('sum')


```
