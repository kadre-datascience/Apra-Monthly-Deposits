```python
import pandas as pd
import plotly
import plotly as py
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
#py.offline.init_notebook_mode(connected = True,)
pio.renderers.default = "notebook"
```

```python
xls = pd.ExcelFile('D:\Kadre\Kadre\APRA\gitaprarepo\Apra-Monthly-Deposits\Data\pillars.xlsx')
TIR = pd.read_excel(xls, sheet_name='TIR', header=0)
xls = pd.ExcelFile('D:\Kadre\Kadre\APRA\gitaprarepo\Apra-Monthly-Deposits\Data\pillars.xlsx')
loss = pd.read_excel(xls, sheet_name='LossR', header=0)
xls = pd.ExcelFile('D:\Kadre\Kadre\APRA\gitaprarepo\Apra-Monthly-Deposits\Data\pillars.xlsx')
Limit = pd.read_excel(xls, sheet_name='Limit_Ut', header=0)
```

```python
tirflip = TIR.transpose().reset_index()
tirflip.columns= tirflip.iloc[0]
tirflip=tirflip[1:]
tirflip.rename(columns={ tirflip.columns[0]: "Time" }, inplace = True)
lossflip = loss.transpose().reset_index()
lossflip.columns= lossflip.iloc[0]
lossflip=lossflip[1:]
lossflip.rename(columns={ lossflip.columns[0]: "Time" }, inplace = True)
Limitflip = Limit.transpose().reset_index()
Limitflip.columns= Limitflip.iloc[0]
Limitlip=Limitflip[1:]
Limitlip.rename(columns={ Limitlip.columns[0]: "Time" }, inplace = True)
```

```python
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tirflip.Time, y=tirflip.ANZ,
                    mode='lines+markers',
                    name='ANZ',line=dict(color='#2B65EC', width=2,
                              )))
fig.add_trace(go.Scatter(x=tirflip.Time, y=tirflip.CBA,
                    mode='lines+markers',
                    name='CBA',line=dict(color='#FFCC00', width=2,
                              )))
fig.add_trace(go.Scatter(x=tirflip.Time, y=tirflip.NAB,
                    mode='lines+markers', name='NAB',line=dict(color='#000000', width=2,
                              )))
fig.add_trace(go.Scatter(x=tirflip.Time, y=tirflip.WBC,
                    mode='lines+markers', name='WBC',line=dict(color='#EB4D2B', width=2,
                              )))         
fig.layout.yaxis.tickformat = ',.1%' 
fig.update_layout(title = 'Imparement Rate of Big 4 (90 DPD)', xaxis_title='Time (Half Years)',
                   yaxis_title='Imparement rate (%)',legend=dict(
        orientation="h",
        title="Banks:",
#        yanchor="bottom",
#        y=1.2,
#        xanchor="left",
 #       x=0.2)
        ))
fig.show()        
```
